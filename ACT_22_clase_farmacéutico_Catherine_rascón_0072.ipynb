```python
class Farmaceutico:

    def __init__(self, nombre, telefono, edad):
        self.nombre = nombre
        self.telefono = telefono
        self.edad = edad

    # Función 1: Barrer
    def barrer(self):
        print(f"{self.nombre} está barriendo el área de farmacia.")

    # Función 2: Trapear
    def trapear(self):
        print(f"{self.nombre} está trapeando el piso para mantener la higiene.")

    # Función 3: Vender Medicamento
    def vende_medicamento(self, medicamento):
        print(f"{self.nombre} ha vendido el medicamento: {medicamento}.")


# Farmacéutico 1
farmaceutico1 = Farmaceutico("Pedro Pérez", "555-0123", 35)

# Farmacéutico 2
farmaceutico2 = Farmaceutico("Ana Martínez", "555-9876", 28)

# --- Ejecución de Funciones y Mensajes ---

print("--- Actividades del Farmacéutico 1 ---")
print(f"Datos: {farmaceutico1.nombre}, Tel: {farmaceutico1.telefono}, Edad: {farmaceutico1.edad}")
farmaceutico1.barrer()
farmaceutico1.vende_medicamento("Paracetamol")
print("------------------------------------------------------------")

print("\n--- Actividades del Farmacéutico 2 ---")
print(f"Datos: {farmaceutico2.nombre}, Tel: {farmaceutico2.telefono}, Edad: {farmaceutico2.edad}")
farmaceutico2.trapear()
farmaceutico2.vende_medicamento("Ibuprofeno")
print("------------------------------------------------------------")
```
